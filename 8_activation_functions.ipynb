# ฟังก์ชันกระตุ้น (Activation functions)

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

สร้าง Grid space และฟังก์ชันสำหรับคำนวณอนุพันธ์

```python
def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)
```

## 1. Sigmoid (Logistic)

```python
z = np.linspace(-7, 7, 200)
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
plt.figure(figsize=(5,4))

# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([-5, 5], [-1, -1], 'k--', linewidth=1)
plt.plot([0, 0], [-2.2, 3.2], 'k-', linewidth=1)
plt.plot([-7,7], [1,1], 'k--', linewidth=1)
# Plot sigmoid
plt.plot(z, sigmoid(z), "b-", linewidth=2, label="Sigmoid")
plt.grid(True)
plt.title("Sigmoid activation function", fontsize=14)
plt.axis([-7, 7, -0.2, 1.2])
plt.show()

plt.figure(figsize=(5,4))
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([-5, 5], [-1, -1], 'k--', linewidth=1)
plt.plot([0, 0], [-2.2, 3.2], 'k-', linewidth=1)
plt.plot(z, derivative(sigmoid, z), "b-", linewidth=2, label="Sigmoid")
plt.grid(True)
plt.title("Derivative", fontsize=14)
plt.axis([-7, 7, -0.2, 1.2])
plt.show()
```

## 3. ReLU

```python
z = np.linspace(-7, 7, 200)
```

```python
def relu(z):
    return np.maximum(0, z)
```

```python
plt.figure(figsize=(5,4))

# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, relu(z), "b-", linewidth=2, label="ReLU")
plt.grid(True)
plt.title("ReLU activation function", fontsize=14)
plt.axis([-5, 5, -0.2, 5])

plt.figure(figsize=(5,4))
# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, derivative(relu, z), "b-", linewidth=2, label="ReLU")
plt.grid(True)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -0.2, 5])
plt.show()
```

## 4. Leaky ReLU

```python
z = np.linspace(-7, 7, 200)
```

```python
def leaky_relu(z, alpha=0.05):
    return np.maximum(alpha * z, z)
```

```python
plt.figure(figsize=(5,4))

# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, leaky_relu(z, 0.05), 'b-', linewidth=2)
plt.plot([-5,5], [0,0], 'k-')
plt.plot([0,0], [-0.5, 5], 'k-')
plt.grid(True)
plt.title("Leaky ReLU activation function", fontsize=14)
plt.axis([-5,5,-0.5, 2])

plt.figure(figsize=(5,4))
# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, derivative(leaky_relu, z), "b-", linewidth=2, label="Leaky ReLU")
plt.grid(True)
plt.title("Derivative", fontsize=14)
plt.axis([-5,5,-0.05, 1.2])
plt.show()
```

## 5. Exponential Linear Unit (ELU)

```python
z = np.linspace(-7, 7, 200)
```

```python
def elu(z, alpha=1):
    return np.where(z < 0, alpha * (np.exp(z) - 1), z)
```

```python
plt.figure(figsize=(5,4))

# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, elu(z), 'b-', linewidth=2)
plt.plot([-5, 5], [0, 0], 'k-')
plt.plot([-5, 5], [-1, -1], 'k--')
plt.plot([0, 0], [-2.2, 3.2], 'k-')
plt.grid(True)
plt.title("ELU activation function", fontsize=14)
plt.axis([-5, 5, -2.2, 3.2])

plt.figure(figsize=(5,4))
# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, derivative(elu, z), "b-", linewidth=2, label="ELU")
plt.grid(True)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -2.2, 3.2])
plt.show()
```

### 6. Scaled Exponential Linear Unit (SELU)

```python
from scipy.special import erfc

# alpha and scale to self normalize with mean 0 and standard deviation 1
# (see equation 14 in the paper):
alpha_0_1 = -np.sqrt(2 / np.pi) / (erfc(1/np.sqrt(2)) * np.exp(1/2) - 1)
scale_0_1 = (1 - erfc(1 / np.sqrt(2)) * np.sqrt(np.e)) * np.sqrt(2 * np.pi) * (2 * erfc(np.sqrt(2))*np.e**2 + np.pi*erfc(1/np.sqrt(2))**2*np.e - 2*(2+np.pi)*erfc(1/np.sqrt(2))*np.sqrt(np.e)+np.pi+2)**(-1/2)
```

```python
alpha_0_1
```

```python
scale_0_1
```

```python
z = np.linspace(-7, 7, 200)
```

```python
def selu(z, scale=scale_0_1, alpha=alpha_0_1):
    return scale * elu(z, alpha)
```

```python
plt.figure(figsize=(5,4))

# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
plt.plot([-5, 5], [-1.758, -1.758], 'k--', linewidth=1)
# Plot
plt.plot(z, selu(z), 'b-', linewidth=2)
plt.grid(True)
plt.title("SELU activation function", fontsize=14)
plt.axis([-5, 5, -2.2, 3.2])

plt.figure(figsize=(5,4))
# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-2.2, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, derivative(selu, z), "b-", linewidth=2, label="SELU")
plt.grid(True)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -2.2, 3.2])
plt.show()
```

## 7. Linear

```python
z = np.linspace(-7, 7, 200)
```

```python
def linear(z,m=1):
    return m*z
```

```python
plt.figure(figsize=(5,4))

# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-5, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, linear(z), "b-", linewidth=2, label="Linear")
plt.grid(True)
plt.title("Linear activation function", fontsize=14)
plt.axis([-5, 5, -5, 5])

plt.figure(figsize=(5,4))
# coordinate 
plt.plot([-7, 7], [0, 0], 'k-', linewidth=1)
plt.plot([0, 0], [-5, 5], 'k-', linewidth=1)
# Plot
plt.plot(z, derivative(linear, z), "b-", linewidth=2, label="Linear")
plt.grid(True)
plt.title("Derivative", fontsize=14)
plt.axis([-5, 5, -5, 5])
plt.show()
```
